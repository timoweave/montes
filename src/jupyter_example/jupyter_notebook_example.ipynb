# Setup
1. need to connect a jupyter notebook file to jupyter server which is running inside docker compose
1. get the jupyter url address
   In the host terminal, run ` docker logs jupyter 2>&1 | grep "?token=" | grep -v "ServerApp" | sed 's/127.0.0.1/jupyter/' | grep jupyter | sed 's/^[ \t]*//' | sed 's/[ \t]*$//' | pbcopy`. it will copy the jupyter url into clipboard and look something like `http://jupyter:8888/lab?token=1ac900d945241791cab3ef51eaadfcd6f37fe2d0162049c4` when press `cmd+v`
1. In vscode, open a jupyter notebook file (with .ipynb extension), 
   run command palette `cmd+shift+p` `> Notebook: Select Notebook Kernel`, please the jupyter url and follow the instruction


```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
plt.plot(np.random.randn(1000).cumsum())
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Sine Curve")
plt.show()
```
